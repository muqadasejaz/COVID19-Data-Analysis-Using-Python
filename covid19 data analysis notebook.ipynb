### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries= list(data.index)
max_infectious_rate= []
for c in countries:
    max_infectious_rate.append(data.loc[c].diff().max())
data["max_infectious_rate"]= max_infectious_rate
```

```python
data.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_data=pd.DataFrame(data["max_infectious_rate"])
```
